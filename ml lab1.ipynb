```python
a = [5,2,8,1,9]
print("original list:", a)
a.append(12)
a.append(7)
print("after append:",a)
a.remove(8)
print("after removing 8:",a)
print("sorted list:",a)
a.reverse()
print("reversed list:",a)
```

```python
A = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("\nMatrix A:")
for i in range(3):
    print(A[i][i])   # prints diagonal elements

B = [
    [9, 8, 7],
    [6, 5, 4],
    [3, 2, 1]
]

# Matrix Addition
addition = []

for i in range(3):
    row = []
    for j in range(3):
        row.append(A[i][j] + B[i][j])
    addition.append(row)

print("\nMatrix Addition:")
for row in addition:
    print(row)

# Matrix Multiplication
x = [
    [1, 2],
    [3, 4]
]

y = [
    [5, 6],
    [7, 8]
]

multiplication = [
    [0, 0],
    [0, 0]
]

for i in range(2):
    for j in range(2):
        for k in range(2):
            multiplication[i][j] += x[i][k] * y[k][j]

print("\nMatrix Multiplication:")
for row in multiplication:
    print(row)
```
